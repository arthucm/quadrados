```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
numeros = range(1, 121)
raiz_digital = []
quadrados = []

for i in numeros:
    quadrado = i ** 2
    quadrados.append(quadrado)
    raiz_d = 9 if quadrado % 9 == 0 and quadrado != 0 else quadrado % 9
    raiz_digital.append(raiz_d)
```

```python
tb_dict = {
    'nr':numeros,
    'raiz_digital':raiz_digital,
    'quadrado':quadrados
}

tb_numeros = pd.DataFrame(tb_dict)
tb_numeros.head()
```

```python
sns.lineplot(data=tb_numeros, x='nr', y='raiz_digital')
```
